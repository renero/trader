# Read and plot konkorde files

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
v1 = pd.read_csv('../output/konkorde_acciona_2018b.csv')
v2 = pd.read_csv('../output/konkorde_acciona_2018.csv')
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(111)

ax1.plot(v1.close, c='black')

ax2 = ax1.twinx()
ax2.set_ylim(-1, 5)
ax2.plot(v1.verde, c='green')
ax2.plot(v1.azul,  c='blue')

plt.show()
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(111)

ax1.plot(v2.close, c='black')

ax2 = ax1.twinx()
ax2.set_ylim(-1, 5)
ax2.plot(v2.verde, c='green')
ax2.plot(v2.azul,  c='blue')

plt.show()
```
