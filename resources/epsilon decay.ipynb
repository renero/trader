# On the importance of $\epsilon$ and $ \epsilon_{min} $ choices

Epsilon controls de ratio between exploration (new random actions to be tested during learning) and exploitation (select only anctions that the algorithm has already learnt). At every step, _epsilon greedy_ algorithm selects between these two options by generating a random number: if its lower than $\epsilon$, then a random action is selected (_exloration_); otherwise, the action is predicted based on the current state of the network (_exploitation_).

At the beginning, we're interested in keeping a very high degree of exploration, and decrease it gradually, as the net learn more and more. This is accomplished by gradually _decaying_ the value of $\epsilon$.

To properly balance between the two desirable states of exploration and exploitation, we must adjust the rate of decay to gradually decrease $\epsilon$ during training, but not too rapidly or too slowly.

Finally, there exist the possibility to preclude the decaying function to completely zero the value of $\epsilon$ by setting an additional parameter called $\epsilon_{min}$. These value is the absolute minimum we're ready to accept, no matter how many iterations we run.

The following plots show possible rates of $\epsilon$ based on the value of the parameter `decay_factor`.

```python
import matplotlib.pyplot as plt
import numpy as np

maxrange=5000
def f(x, decay):
    return x*decay

Y = dict()
for decay in np.arange(0.99, 0.999, 0.001):
    Y[str(decay)] = []
    x = 1.0
    Y[str(decay)].append(x)
    for _ in range(maxrange):
        x = f(x, decay)
        Y[str(decay)].append(x)

plt.figure(figsize=(12,8))
plt.title('$\epsilon$ decay')
for d in Y.keys():
    plt.plot(Y[d], label='$\epsilon$ decay={}'.format(d))
plt.legend()
plt.show()
```
