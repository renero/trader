## Generic function

```python
def plot_data_and_konkorde(file, combine=False):
    data = pd.read_csv(file)
    fig = plt.figure(figsize=(12,8))
    ax1 = fig.add_subplot(111)

    ax1.plot(data.actual, c='black')

    ax2 = ax1.twinx()
    ax2.set_ylim(-2, 10)
    if combine is False:
        ax2.plot(data.verde, c='green', linewidth=0.5)
        ax2.plot(data.azul,  c='blue', linewidth=0.5)
    else:
        ax2.set_ylim(-2, +10.)
        ax2.axhline(0, color='black', alpha=0.3)
        ax2.fill_between(range(data.verde.shape[0]), 0, data.verde+data.azul,
                         color='green', alpha=0.3)
        # ax2.plot(data.verde+data.azul, c='green', linewidth=0.5)
    ax2.axhline(0, c='r', linewidth=0.5)

    plt.show()
```

```python
plot_data_and_konkorde('../output/BEARDAX/forecast_DAX_highlow.csv', combine=True)
```
